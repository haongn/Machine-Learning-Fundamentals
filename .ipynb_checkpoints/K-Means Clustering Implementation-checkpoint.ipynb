Notebook này trình bày cách xây dựng thuật toán phân cụm K-Means (K-Means Clustering).

Hàm cdist trong scipy.
spatial.distance được dùng để tính khoảng cách giữa các cặp điểm trong hai tập
hợp một cách hiệu quả (việc xây dựng hàm số này không sử dụng thư viện đã được thảo luận kỹ trong Chương 9) (sử dụng vectorization + broadcasting).

```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns; sns.set()
from scipy.spatial.distance import cdist
import random 

np.random.seed(11)
```

\begin{align}
\mathbf{X} &= [\mathbf{x}_1 \quad \mathbf{x}_2 \quad ... \quad \mathbf{x}_N] \in \mathbb{R}^{d \times N} \quad \text{các điểm dl là các vector cột, có N điểm dl} \\
\mathbf{M} &= [\mathbf{m}_1 \quad \mathbf{m}_2 \quad ... \quad \mathbf{m}_k] \in \mathbb{R}^{d \times K} \quad \text{MT tạo bởi K tâm cụm} \\
\mathbf{Y} &= 
\begin{bmatrix}
\mathbf{y}_1 \\
\mathbf{y}_2 \\
\cdots \\
\mathbf{y}_N
\end{bmatrix}
\in \mathbb{R}^{N \times d}
\quad \text{là MT nhãn dán cho tất cả các điểm DL (hàng y_i là nhãn cho điểm DL x_i)}
\end{align}

Trước hết, ta tạo tâm cụm và dữ liệu cho
từng cụm bằng cách lấy mẫu theo phân phối chuẩn có kỳ vọng là tâm của cụm
đó và ma trận hiệp phương sai là ma trận đơn vị (tức các biến $\mathbf{x}_i$ đôi một không tương quan với nhau).

Dữ liệu được tạo bằng cách lấy ngẫu nhiên 500 điểm cho mỗi cụm theo phân phối
chuẩn có kỳ vọng lần lượt là (2, 2), (8, 3) và (3, 6); ma trận hiệp phương sai
giống nhau và là ma trận đơn vị.

```python
means = [[2, 2], 
         [8, 3], 
         [3, 6]]
cov = [[1, 0], 
       [0, 1]]

N = 500     # số điểm DL 

X0 = np.random.multivariate_normal(means[0], cov, N)  # sinh N điểm DL hai chiều (MT N x 2)
X1 = np.random.multivariate_normal(means[1], cov, N)  # sinh N điểm DL hai chiều (MT N x 2)
X2 = np.random.multivariate_normal(means[2], cov, N)  # sinh N điểm DL hai chiều (MT N x 2)

X = np.concatenate((X0, X1, X2), axis = 0)    # tạo ma trận đặc trưng (nối theo chiều dọc) (MT 3N x 2)
```

```python
K = 3   # 3 clusters 
original_label = np.asarray([0]*N + [1]*N + [2]*N).T     # list concatenation (row vector)
```

Chúng ta cần một hàm kmeans_display để hiển thị dữ liệu. Sau đó hiển thị dữ liệu theo nhãn ban đầu:

```python
def kmeans_display(X, label): 
    """Hiện thị dữ liệu theo nhãn ban đầu (label).
    X: feature matrix 
    label: an array of exact label (0, 1, ..., k)
    """
    
    X0 = X[label == 0, :]    # chọn những phần tử mà có index của mảng label == 0 (label == 0: một mảng True False)
    X1 = X[label == 1, :]    # chọn những phần tử mà có index của mảng label == 1 
    X2 = X[label == 2, :]    # chọn những phần tử mà có index của mảng label == 2 
    
    plt.figure(figsize = (10, 6))
    plt.plot(X0[:, 0], X0[:, 1], 'b^', markersize = 4, alpha = .8)   # scatter plot of cluster 1 
    plt.plot(X1[:, 0], X1[:, 1], 'go', markersize = 4, alpha = .8)   # scatter plot of cluster 2 
    plt.plot(X2[:, 0], X2[:, 1], 'rs', markersize = 4, alpha = .8)   # scatter plot of cluster 3 

    plt.axis('equal')
    plt.plot()
    plt.show()
```

```python
kmeans_display(X, original_label)    
```

Trước khi viết thuật toán chính phân cụm K-means, ta cần một số hàm con sau: \
a. **kmeans_init_centroids** khởi tạo các tâm cụm. \
b. **kmeans_assign_labels** tìm nhãn mới cho các điểm khi biết các tâm cụm. \
c. **kmeans_update_centroids** cập nhật các tâm cụm khi biết nhãn của từng điểm. \
d. **has_converged** kiểm tra điều kiện dừng của thuật toán.

```python
def kmeans_init_centroids(X, k): 
    """randomly pick k rows of X as initial centroids.
    X: feature matrix
    k: number of centroids 
    """
    return X[np.random.choice(X.shape[0], size = k, replace = False)]   # trả về một ma trận 
```

```python
def kmeans_assign_labels(X, centroids): 
    """calculate pairwise distances between data and centroids.
    """
    D = cdist(X, centroids)            # distance matrix, ith row is distance of ith point to k centroids 
    # return index of the closet centroid 
    return np.argmin(D, axis = 1)      # trả về index của phần tử min của từng hàng (cũng là cụm nó thuộc về) 
```

```python
def has_converged(centroids, new_centroids): 
    """Return True if two sets of centroids are the same.
    """
    return (set([tuple(a) for a in centroids]) == set([tuple(a) for a in new_centroids]))
```

```python
def kmeans_update_centroids(X, labels, K): 
    """Update new centroids.
    X: feature matrix 
    labels: array of labels 
    K: number of centroids 
    """
    
    centroids = np.zeros((K, X.shape[1]))    # create a zero matrix of size K x 2 
    for k in range(K): 
        # collect all points that are assigned to the k-th cluster 
        Xk = X[labels == k, :]               # return a matrix 
        centroids[k, :] = np.mean(Xk, axis = 0)    # take average of k-th cluster 
    
    return centroids 
```

Hàm chính của phân cụm K-means:

```python
def kmeans(X, K): 
    """X: features matrix 
    K: number of clusters
    """
    centroids = [kmeans_init_centroids(X, K)]    # randomly pick k centroids 
    labels = []                                  # labels of all points 
    it = 0                                       # number of loops 
    while True:
        labels.append(kmeans_assign_labels(X, centroids[-1]))          # insert new row (new labels row)
        new_centroids = kmeans_update_centroids(X, labels[-1], K)      
        if has_converged(centroids[-1], new_centroids): 
            break 
        
        centroids.append(new_centroids)         # append new row (new_centroids)
        it += 1 
        
    return (centroids, labels, it)
```

Áp dụng thuật toán vừa viết vào dữ liệu ban đầu và hiển thị kết quả cuối cùng:

```python
centroids, labels, it = kmeans(X, K)

print('Centers found by our algorithm:\n', centroids[-1])
print('Number of loops: %d' % (it))

kmeans_display(X, labels[-1])
```

### So sánh kết quả với hàm của thư viện Scikit-learn

```python
from sklearn.cluster import KMeans 

kmeans = KMeans(n_clusters = 3, random_state = 0).fit(X)

print('Center found by scikit-learn: ')
print(kmeans.cluster_centers_)

pred_label = kmeans.predict(X)

kmeans_display(X, pred_label)
```
